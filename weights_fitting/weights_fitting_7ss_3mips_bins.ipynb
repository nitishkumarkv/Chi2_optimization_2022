```python
columns=["true_ene","w1","w2","w3","w1_err","w2_err","w3_err"]

file_folder = "10M_aug_4_10bins/Results_new_chi2_10_bins/"



#save_folder = "weights_ecal_ene_frac_10M/"
save_folder = "/eos/user/n/nkasarag/www/folder/Chi_square_optimization/new_chi2_10M/10M_aug_4_10bins/fitted_weights/mipsInEE_weights_10M_fitted/"
```

```python
mips_bin1 = pd.read_csv(file_folder+"/chi2_opt_mipsInEE_hcal_ene_fracFact_0.000000_to_0.400000_flatEn.txt", sep="\t", header=None)
mips_bin1.columns=columns

mips_bin2 = pd.read_csv(file_folder+"/chi2_opt_mipsInEE_hcal_ene_fracFact_0.400000_to_0.700000_flatEn.txt", sep="\t", header=None)
mips_bin2.columns=columns

mips_bin3 = pd.read_csv(file_folder+"/chi2_opt_mipsInEE_hcal_ene_fracFact_0.700000_to_1.010000_flatEn.txt", sep="\t", header=None)
mips_bin3.columns=columns

"""mips_bin4 = pd.read_csv(file_folder+"/chi2_opt_mipsInEE_hcal_ene_fracFact_0.750000_to_0.900000_flatEn.txt", sep="\t", header=None)
mips_bin4.columns=columns

mips_bin5 = pd.read_csv(file_folder+"/chi2_opt_mipsInEE_hcal_ene_fracFact_0.900000_to_1.010000_flatEn.txt", sep="\t", header=None)
mips_bin5.columns=columns

mips_bin5.head()"""
```

```python
name = 'w1_bin1_mipsInEE'

c1 = TCanvas(name, name, 900, 700 )
 
c1.Range(-60.25,-0.625,562.25,0.625)
c1.SetGrid()
c1.GetFrame().SetFillColor( 0 )
c1.GetFrame().SetBorderSize( 2 )
c1.GetFrame().SetBorderMode( 0 )
df = mips_bin1

n = df.true_ene.size
x  = df.true_ene.values.astype(np.float64)
y  = df.w1.values.astype(np.float64)
ex = np.zeros(x.size).astype(np.float64)
ey = df.w1_err.values.astype(np.float64)

title = "0< FH_{frac}< 0.4 (MipsInEE); Energy (GeV); w1"

f1 = ROOT.TF1("f1","sqrt([0]*[0]+[1]/x)+[2]/x+[3]",10,350)

gr = TGraphErrors( n, x, y, ex, ey)
gr.SetTitle( title )
gr.SetMarkerColor( 4 )
gr.SetMarkerStyle( 8 )


Fit = gr.Fit("f1", "S", "",10,350)

param[0] = Fit.Get().Parameter(0)
param[1] = Fit.Get().Parameter(1)
param[2] = Fit.Get().Parameter(2)
param[3] = Fit.Get().Parameter(3)

print(Fit.Get().Parameter(0))
print(Fit.Get().Parameter(1))
print(Fit.Get().Parameter(2))
print(Fit.Get().Parameter(3))

gr.Draw( 'ALP' )
ROOT.gStyle.SetOptFit(1)
#c1.Update()
c1.Draw()

c1.SaveAs(save_folder+name+".png")
```
