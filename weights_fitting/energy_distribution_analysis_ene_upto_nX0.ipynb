```python
lst = ['trueBeamEnergy', 'rechit_shower_start_layer', 'rechit_energyPerLayer','ahc_energyPerLayer']

df = data["pion_variables_v1"].pandas.df(lst, flatten=False, entrystart=0, entrystop=50)
df.head()
```

```python
df.ahc_energyPerLayer.values[0].shape
```

```python
def tot_ene_ecal(df_):
    return np.sum(df_.rechit_energyPerLayer[:28])

def tot_ene_hcal(df_):
    return np.sum(df_.rechit_energyPerLayer[28:])

def tot_ene_ahcal(df_):
    return np.sum(df_.ahc_energyPerLayer)
```

```python
df["ene_ecal"] = df.apply(tot_ene_ecal,axis=1)
df["ene_hcal"] = df.apply(tot_ene_hcal,axis=1)
df["ene_ahcal"] = df.apply(tot_ene_ahcal,axis=1)
df.head()
```
