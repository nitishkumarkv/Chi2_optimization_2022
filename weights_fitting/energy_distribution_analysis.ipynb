```python
def ene_upto_nX0(df_, l):
    
    ss_lay_idx = df_.rechit_shower_start_layer-1
    max_lay_idx = np.argmin(abs(rad_len-(rad_len[ss_lay_idx]+l)))
    
    return np.sum(df_.ene_per_lay_all_GeV[ss_lay_idx: max_lay_idx+1])

ahcal_lay = np.array([43,47,51,55,59,63,67,71,75,79])

def ene_per_lay_all_GeV(df_):
    
    ahcal_ene_lay = np.array([df_.ahc_energyPerLayer[ahcal_lay[i]-1-40] for i in range(10)])
    
    return np.concatenate((df_.rechit_energyPerLayer*fix_wt_lay, ahcal_ene_lay*0.12504))

fix_wt_lay = (np.arange(1,41)<=28)*0.0105 + (np.arange(1,41)>28)*0.0812
fix_wt_lay
```

```python
df["ene_per_lay_all_GeV"] = df.apply(ene_per_lay_all_GeV, axis=1)

#df["ene_upto_7_X0"] = df.apply(ene_upto_nX0, l=7, axis=1)
df["ene_upto_9_X0"] = df.apply(ene_upto_nX0, l=9, axis=1)
df["ene_upto_12_X0"] = df.apply(ene_upto_nX0, l=12, axis=1)
#df["ene_upto_15_X0"] = df.apply(ene_upto_nX0, l=15, axis=1)
#df["ene_upto_18_X0"] = df.apply(ene_upto_nX0, l=18, axis=1)
df.head()
```
