```python
lst = ['trueBeamEnergy', 'rechit_shower_start_layer', 'rechit_energyPerLayer','ahc_energyPerLayer']

df = data["pion_variables_v1"].pandas.df(lst, flatten=False)#, entrystart=0, entrystop=50)
df.head()
```

```python
def tot_ene_ecal(df_):
    return np.sum(df_.rechit_energyPerLayer[:28])

def tot_ene_hcal(df_):
    return np.sum(df_.rechit_energyPerLayer[28:])

def tot_ene_ahcal(df_):
    return np.sum(df_.ahc_energyPerLayer)
```

```python
df["ene_ecal"] = df.apply(tot_ene_ecal,axis=1)
df["ene_hcal"] = df.apply(tot_ene_hcal,axis=1)
df["ene_ahcal"] = df.apply(tot_ene_ahcal,axis=1)
df.head()
```

```python
trueE = df.trueBeamEnergy.values
ecal_ene = df.ene_ecal.values
hcal_ene = df.ene_hcal.values
ahcal_ene = df.ene_ahcal.values
```

```python
tot_ene = ecal_ene*0.0105 + hcal_ene*0.0789 + ahcal_ene*0.0316
ecal_ene_fraction = (ecal_ene*0.0105)/tot_ene
hcal_ene_fraction = (hcal_ene*0.0789)/tot_ene
ahcal_ene_fraction = (ahcal_ene*0.0316)/tot_ene

df["ecal_ene_fraction"] = ecal_ene_fraction
df["hcal_ene_fraction"] = hcal_ene_fraction
df["ahcal_ene_fraction"] = ahcal_ene_fraction

df.head()
```
