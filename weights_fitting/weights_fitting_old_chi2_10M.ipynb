```python
from ROOT import TCanvas, TGraphErrors
from ROOT import gROOT
import uproot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import ROOT
%jsroot on
```

```python
columns=["true_ene","w1","w2","w3","w1_err","w2_err","w3_err"]

file_folder = "old_chi2_10M/Results/"



#save_folder = "weights_ecal_ene_frac_10M/"
save_folder = "/eos/user/n/nkasarag/www/folder/Chi_square_optimization/old_chi2_10M/Fitted_weights/"
```

```python
H_hadrons = pd.read_csv(file_folder+"chi2_calibFact_H_hadrons_flatEn_1_v1.txt", sep="\t", header=None)
H_hadrons.columns=columns

EH_hadrons = pd.read_csv(file_folder+"chi2_calibFact_EH_hadrons_flatEn_1_v1.txt", sep="\t", header=None)
EH_hadrons.columns=columns

EH_hadrons.head()
```

```python
name = 'H_w1'

c1 = TCanvas(name, name, 900, 700 )
 
c1.Range(-60.25,-0.625,562.25,0.625)
c1.SetGrid()
c1.GetFrame().SetFillColor( 0 )
c1.GetFrame().SetBorderSize( 2 )
c1.GetFrame().SetBorderMode( 0 )
df = H_hadrons

n = df.true_ene.size
x  = df.true_ene.values.astype(np.float64)
y  = df.w1.values.astype(np.float64)
ex = np.zeros(x.size).astype(np.float64)
ey = df.w1_err.values.astype(np.float64)

title = "H_Hadron; Energy (GeV); w1"

f1 = ROOT.TF1("f1","sqrt([0]*[0]+[1]/x)+[2]/x+[3]",10,350)

gr = TGraphErrors( n, x, y, ex, ey)
gr.SetTitle( title )
gr.SetMarkerColor( 4 )
gr.SetMarkerStyle( 8 )


Fit = gr.Fit("f1", "S", "",10,350)
print(Fit.Get().Parameter(0))
print(Fit.Get().Parameter(1))
print(Fit.Get().Parameter(2))
print(Fit.Get().Parameter(3))

gr.Draw( 'ALP' )
ROOT.gStyle.SetOptFit(1)
#c1.Update()
c1.Draw()

#c1.SaveAs(save_folder+name+".png")
```
